```python
data.drop_duplicates(inplace=True)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
# Extract month and year from the date
data['month'] = data['Order Date'].dt.month
data['year'] = data['Order Date'].dt.year
```

```python
# Aggregate data for month-wise, year-wise, and yearly_month-wise analysis
monthly_sales = data.groupby(['year', 'month']).agg({'Total Cost': 'sum'}).reset_index()
yearly_sales = data.groupby('year').agg({'Total Cost': 'sum'}).reset_index()
yearly_monthly_sales = data.pivot_table(index='month', columns='year', values='Total Cost', aggfunc='sum').reset_index()
```

```python
# Month-wise Sales Trend
plt.figure(figsize=(10, 6))
monthly_sales.groupby('month')['Total Cost'].mean().plot(kind='line')
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Year-wise Sales Trend
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales['year'], yearly_sales['Total Cost'], marker='o')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Yearly Month-wise Sales Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(yearly_monthly_sales.set_index('month'), annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
# Total Sales
total_sales = data['Total Cost'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Average Sales per Month/Year
avg_sales_per_month = data.groupby('month')['Total Cost'].mean()
avg_sales_per_year = data.groupby('year')['Total Cost'].mean()
print("Average Sales per Month:")
print(avg_sales_per_month)
print("\nAverage Sales per Year:")
print(avg_sales_per_year)
```

```python
# Sales Growth Rate (Year-over-Year)
yearly_sales['growth_rate'] = yearly_sales['Total Cost'].pct_change() * 100
print("Year-over-Year Sales Growth Rate:")
print(yearly_sales)
```

```python
# Correlation between sales and discount (if discount data is available)
if 'discount' in data.columns:
    correlation = data[['Total Cost', 'discount']].corr()
    print(f"Correlation between Sales and Discount:\n{correlation}")
else  :
    print("Discount data is not available in the dataset.")
```
